4. Using the given x and y data, generate a 3D surface plot to visualize the function 


```python
import plotly.graph_objects as go


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)


fig = go.Figure(data=[go.Surface(z=df.pivot_table(index='Y', columns='X', values='Z').values)])
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x-axis), and bubble size proportional to the population
    

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df['GDP'],
    y=df['Population'],
    mode='markers',
    marker=dict(
        size=df['Population'],  
        sizemode='diameter',  
        sizeref=2.*max(df['Population'])/(15.**2),  
        sizemin=4 
    ),
    text=df['Country']  
))

fig.update_layout(
    title='Bubble Chart of Population and GDP',
    xaxis_title='GDP',
    yaxis_title='Population'
)

fig.show()
```

## BOKEH ASSIGNMENT:

1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x=np.linspace(0,10,100)
y= np.sin(x)

p= figure(title= "Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x,y, legend_label= "sin(x)", line_width=2)
show(p)
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
import random
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

x= [random.random() for _ in range(100)]
y=[random.random() for _ in range(100)]
sizes= [random.randint(10,30) for _ in range(100)]
colors = ["#%02x%02x%02x" % (int(r), int(g), int(b)) for r, g, b in zip(np.random.randint(0, 256, 100), np.random.randint(0, 256, 100), np.random.randint(0, 256, 100))]

source= ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors= colors))

p= figure(title= "Scatter Plot with Random Sizes and Colors")
p.scatter("x", "y", size="sizes", color = "colors", alpha=0.6, source = source)
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

    

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data

    

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p=figure(title= "Histogram of Random Data", x_axis_label= "Value", y_axis_label= "Frequency")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

show(p)
```

5. Create a Bokeh heatmap using the provided dataset

    

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))

p.add_layout(color_bar, 'right')

show(p)
```
